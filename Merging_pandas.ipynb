```python
#merging dataframes in pandas (join, merge, concat)
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\hp pavilion\Downloads\LOTR.csv") #reading the csv file
df
```

```python
df2 = pd.read_csv(r"C:\Users\hp pavilion\Downloads\LOTR 2.csv") 
df2
```

```python
df.merge(df2, how = 'outer')
```

```python
#joining dataframes using the join method
df3 = df.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_left', rsuffix = '_right', how = 'outer')
df3
```

```python
df4 = pd.concat([df, df2], axis=1) #concatenating the two dataframes if the axis is 0 it will concatenate the dataframes vertically and if it is 1 it will concatenate the dataframes horizontally
df4
```
