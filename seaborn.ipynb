```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\hp pavilion\Downloads\real_estate.csv")
df
```

```python
housing_filter = df[df['state'].isin(['Puerto Rico', 'Massachusetts', 'New York'])]
sns.barplot(x='state', y = 'price', data = housing_filter, estimator = 'median')
sns.set_palette("grey")
sns.set_context("notebook")
plt.title('Median Housing Prices in Puerto Rico, Massachusetts, and New York')
```

```python
sns.set_palette("muted")
sns.set_context("notebook")
housing_filter = df[df['house_size'] < 10000]
sns.regplot(x= 'house_size', y = 'price', data = housing_filter, line_kws={'color': 'red'})
plt.title('House Size vs. Price')

plt.ticklabel_format(style='plain', axis = 'y')
plt.show()
```

```python
df_2 = pd.read_csv(r"C:\Users\hp pavilion\Downloads\smoking.csv")
df_2
```

```python
smoking = df_2[df_2['smoke'] == 'Yes']  #filtering the data to only include smokers

male_smokers = smoking[smoking['gender'] == 'Male']  #filtering the data to only include male
female_smokers = smoking[smoking['gender'] == 'Female']  #filtering the data to only include female
```

```python
fig , ax = plt.subplots(1, 2, figsize = (12, 6)) 
sns.histplot(female_smokers['age'], bins = 20, ax = ax[0], color = 'pink')
sns.histplot(male_smokers['age'], bins=20, ax = ax[1], color = 'blue')
ax[0].set_title('female_smokers - age distribution')
ax[1].set_title('Male_smokers - age distribution')

plt.tight_layout()
plt.show()
```

```python
housing_df = pd.read_csv(r"C:\Users\hp pavilion\Downloads\real_estate (1).csv", encoding = 'latin1')
housing_df
```

```python
housing_filter = housing_df.drop('zip_code', axis = 1)

housing_filter
```

```python
numeric_housing_filter = housing_filter.select_dtypes(include=[np.number])
correlation = numeric_housing_filter.corr()
correlation
```

```python
sns.heatmap(correlation, annot= True, cmap='coolwarm') #annot = True to display the correlation values on the heatmap
```

```python
pearson_corr = numeric_housing_filter.corr(method='pearson')
pearson_corr
sns.heatmap(pearson_corr, annot= True, cmap='coolwarm')
```

```python
spear_corr = numeric_housing_filter.corr(method='spearman')
spear_corr
```

```python
sns.heatmap(spear_corr, annot= True, cmap='coolwarm')
```

```python
kende_corr = numeric_housing_filter.corr(method='kendall')
sns.heatmap(kende_corr, annot= True, cmap='coolwarm')
```

```python
fig, ax = plt.subplots(3,1, figsize = (8, 6))
sns.heatmap(pearson_corr, annot= True, cmap= 'coolwarm', ax = ax[0])
ax[0].set_title('pearson_corr')
sns.heatmap(spear_corr, annot= True, cmap= 'coolwarm', ax = ax[1])
ax[1].set_title('spear_corr')
sns.heatmap(kende_corr, annot= True, cmap= 'coolwarm', ax = ax[2])
ax[2].set_title('kende_corr')

plt.tight_layout()
```
